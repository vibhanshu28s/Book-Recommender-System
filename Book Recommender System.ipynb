#
## Popularity Based Book Recommender System 

```python
books_with_ratings=ratings.merge(books,on="ISBN")
```

```python
books_with_ratings
```

```python
books_with_ratings['Book-Rating']
```

```python
num_rating_df=books_with_ratings.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = books_with_ratings.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df)
```

```python
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_df
```

#
## Collaborative Filtering Based Recommander System

```python
X=books_with_ratings.groupby('User-ID').count()['Book-Rating']>200
read_book=X[X].index
```

```python
filtered_rating=books_with_ratings[books_with_ratings['User-ID'].isin(read_book)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books=y[y].index
```

```python
final_ratings=books_with_ratings[books_with_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Harry Potter and the Chamber of Secrets (Book 2)')
```
